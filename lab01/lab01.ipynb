```python
import numpy as np

def mystats(data):
    wynik = np.median(data)
    return wynik

def trimmed_mean(data, p):
    arr = np.sort(np.array(data))
    n = len(arr)

    if isinstance(p, float) and 0 <= p < 0.5:
        odciecie = int(n*p)
    else:
        odciecie = int(p)

    arr = arr[odciecie:-odciecie]

    wynik2 = np.mean(arr)
    return wynik2



```

```python
import pandas as pd

kolumny = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
df = pd.read_csv('iris.data', names=kolumny)

cechy = df.drop(columns=['class'])
print("globalna wartość średnia")
print(cechy.mean())
print("globalna wartość mediana")
print(cechy.median())
print("globalne odchylenie standardowe")
print(cechy.std())

print("statystyki dla kazdego irysa oddzielnie")
statystyki_grupowe = df.groupby('class').agg(['mean', 'median', 'std'])

display(statystyki_grupowe)
```

```python
import pandas as pd

df_red = pd.read_csv('winequality-red.csv', sep=';')
df_white = pd.read_csv('winequality-white.csv', sep=';')

df_red['color'] = 'red'
df_white['color'] = 'white'

df_wine = pd.concat([df_red, df_white], ignore_index=True)

df_wine.columns = df_wine.columns.str.replace(' ', '_')

print("zadanie 4")
cechy_bez_targetu = df_wine.drop(columns=['quality', 'color'])

print("\nwariancja cech posortowana rosnąco:")
wariancja_cech = cechy_bez_targetu.var().sort_values(ascending=True)
print(wariancja_cech)

print("\nwariancja cech dla każdej unikalnej wartości cechy quality:")
wariancja_po_quality = df_wine.drop(columns=['color']).groupby('quality').var()
display(wariancja_po_quality)

print("\nuśredniona wariancja ze wszystkich cech dla każdej wartości cechy quality':")
srednia_wariancja_quality = wariancja_po_quality.mean(axis=1)
print(srednia_wariancja_quality)

print("\nodchylenie standardowe z podziałem na kolor wina:")
odchylenie_po_kolorze = df_wine.drop(columns=['quality']).groupby('color').std()
display(odchylenie_po_kolorze)

print("\nzadanie 5")
kwartyle = cechy_bez_targetu.quantile([0.25, 0.50, 0.75])
display(kwartyle)

print("\nzadanie 6")
percentyl_95 = df_wine['residual_sugar'].quantile(0.95)
powyzej_95 = df_wine[df_wine['residual_sugar'] > percentyl_95]
display(powyzej_95)

```
